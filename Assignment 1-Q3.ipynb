```python
import pandas as pd
df = pd.read_html(url, flavor='html5lib')[0]
df
```

# Ignore cells with a borough that is Not assigned

```python
new_df = df.query("Borough != 'Not assigned'")
new_df = new_df.reset_index(drop=True)
new_df
```

# If Postal code is listed twice and has two neighborhoods, these two rows will be combined into one row with the neighborhoods separated with a comma

```python
df_grouped = new_df.groupby(['Postal Code','Borough'], as_index=False).agg(lambda x: ','.join(x))
df_grouped
```

# If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
NA = df_grouped['Neighbourhood'] == "Not assigned"
df_grouped.loc[NA, 'Neighbourhood'] = df_grouped.loc[NA, 'Borough']
df_grouped
```

# Get the latitude and the longitude coordinates of each neighborhood

```python
import pandas as pd
import os
coordenate = pd.read_csv("C:/Users/admin/Desktop/Geospatial_Coordinates.csv")
coordenate
```

```python
neighborhood = pd.merge(df_grouped, coordenate, on='Postal Code', how='inner')
neighborhood
```

# Use geopy library to get the latitude and longitude values of Toronto.

```python
address = 'Toronto'
geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
map_Toronto
```

```python
# add markers to map
for lat, lng, label in zip(neighborhood['Latitude'], neighborhood['Longitude'], neighborhood['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
